```python
#Using Spacy

import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("Hello there. I am learning NLP! Tokenization is easy.")

#tokenization
for token in doc:
    print(token.text)  
```

```python
#Using Spacy

import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("Hello there. I am learning NLP! Tokenization is easy.")

#tokenization
for token in doc.sents:
    print(token.text)
```
