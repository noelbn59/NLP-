```python
# Count Vectorization

from sklearn.feature_extraction.text import CountVectorizer

texts = [
    "I love python",
    "python is great"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```
