```python
# PorterStemmer

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "runner", "flies", "easily", "fairly"]

for w in words:
    print(w, "→", stemmer.stem(w))
```

```python
# PorterStemmer

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["playing", "player", "played", "plays"]

for w in words:
    print(w, "→", stemmer.stem(w))
```

```python
# SPACY STEMMING(PorterStemmer)


import spacy
from nltk.stem import PorterStemmer

nlp = spacy.load("en_core_web_sm")

stemmer = PorterStemmer()

doc = nlp("The children were happily running and playing in the garden")

for token in doc:
    print(token.text, "->", stemmer.stem(token.text))
```
